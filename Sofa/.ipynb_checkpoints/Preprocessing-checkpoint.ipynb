```python
from docx import Document
import re
```

```python
def make_paragraphs_and_list_of_titles(document):
    """Делаем первоначальный пул параграфов"""
    """На вход документ в формате Document из библиотеки docx, на выходе - параграфы"""
    pattern_par = re.compile('\(.+\)')
    paragraphs = []
    titles = []
    temp1 = []
    temp2 = []
    
    #Убираем все пустые параграфы и параграфы по типу (в ред.)
    for paragraph in document.paragraphs:
        p = paragraph.text
        if p != '' and re.match(pattern_par,p) is None:
            paragraphs.append(p)
            
        if paragraph.style.name == 'ConsPlusTitle':
            if paragraph.text != '':
                titles.append(paragraph.text)
    
    #Собираем все титулы, чтобы обрезать его сверху до Главы 1
    for i in range(len(titles)):
        for j in range(len(paragraphs)):
            if titles[i] == paragraphs[j]:
                temp1.append(j)
                if paragraphs[j][:5] == 'Глава' or paragraphs[j][:5] == 'Стать':
                    temp2.append(j)   
    paragraphs = paragraphs[temp2[0]:]
    
    #В конце параграфов находим место подписи и обрезаем параграфы снизу
    for i in range(len(paragraphs)):
        if paragraphs[i] == 'Президент' or paragraphs[i] == 'Председатель':
            n = i
    paragraphs = paragraphs[:n] 
    
    return paragraphs
```

```python
def first_steps(paragraphs):
    """Обработка на первом этапе, чтобы прицепить все подпункты к пунктам,
    а пункты к частям"""
    """На вход параграфы, на выходе - обработанные параграфы"""
    #Вылавливаем подпункты
    pattern_lessp = re.compile('\d+.\d+\)')
    lesspuncts = []
    for i in range(len(paragraphs)):
        if re.match(pattern_lessp, paragraphs[i]) is not None:
            lesspuncts.append(i)
            
    #Формируем пулы подпунктов     
    ness_lesspuncts = []
    for index in range(1,len(lesspuncts)):
        if lesspuncts[index]-lesspuncts[index-1] !=1:
            ness_lesspuncts.append(index)
            
    #Считаем длину каждого из пулов подпунктов
    lesspuncts2 = lesspuncts
    lessness_2 = [int(ness_lesspuncts[i])-int(ness_lesspuncts[i-1])\
                  for i in range(1,len(ness_lesspuncts))]
    
    #Начинаем пришивать подпункты к пунктам
    if len(lesspuncts) != 1:
        new1_list = lesspuncts[:ness_lesspuncts[0]]
        for i in new1_list:
            first = paragraphs[int(new1_list[0]-1)]
            paragraphs[i] = first+' '+paragraphs[i][4:]   
        lesspuncts = lesspuncts[ness_lesspuncts[0]:]
        
        for i in range(len(lessness_2)):
            temp_list = lesspuncts[:lessness_2[i]]
            for j in temp_list:
                first = paragraphs[int(temp_list[0]-1)]
                paragraphs[j] = first+' '+paragraphs[j][4:]
            lesspuncts = lesspuncts[lessness_2[i]:]
        
        for i in lesspuncts:
            first = paragraphs[int(lesspuncts[0]-1)]
            paragraphs[i] = first+' '+paragraphs[i][4:]
        
    else:
        first = paragraphs[int(lesspuncts[0]-1)]
        paragraphs[int(lesspuncts[0])] = first+' '+paragraphs[int(lesspuncts[0])][4:]
        
        
    #Вылавливаем подпункты     
    pattern = re.compile('\d+\)')
    puncts = []
    for i in range(len(paragraphs)):
        if re.match(pattern, paragraphs[i]) is not None:
            puncts.append(i)
            
    #Формируем пулы пунктов         
    ness = []
    for index in range(1,len(puncts)):
        if puncts[index]-puncts[index-1] !=1:
            ness.append(index)
    puncts2 = puncts
    
    ness3 = []
    ness3.append(0)
    for i in range(len(ness)):
        ness3.append(ness[i])
        
    #Считаем длину каждого из пулов подпунктов    
    ness_2 = [int(ness[i])-int(ness[i-1]) for i in range(1,len(ness))]
    
    #Начинаем пришивать пункты к частям
    new_list = puncts[:ness[0]]
    for i in new_list:
        first = paragraphs[int(new_list[0]-1)]
        paragraphs[i] = first+' '+paragraphs[i][3:]   
    puncts = puncts[ness[0]:]
    
    for i in range(len(ness_2)):
        temp_list = puncts[:ness_2[i]]
        for j in temp_list:
            first = paragraphs[int(temp_list[0]-1)]
            paragraphs[j] = first+' '+paragraphs[j][3:]
        puncts = puncts[ness_2[i]:]
    
    for i in puncts:
        first = paragraphs[int(puncts[0]-1)]
        paragraphs[i] = first+' '+paragraphs[i][3:]
        
    #Части без пунктов, которые мы пришивали ранее, удаляем   
    for i in range(len(ness3)):
        paragraphs[int(puncts2[ness3[i]])-1] = ''
    
    #Чистим параграфы от пропусков 
    paragraphs_new = []
    for i in paragraphs:
        if i != '':
            paragraphs_new.append(i)
            
    return paragraphs_new
```

```python
def list_of_titles(document):
    """Делаем список титулов из всего документа (с учетом даже первых)"""
    """На вход документ в формате Document из библиотеки docx, на выходе - список титулов"""
    titles = []
    for paragraph in document.paragraphs:
        if paragraph.style.name == 'ConsPlusTitle':
            if paragraph.text != '':
                titles.append(paragraph.text)
    return titles
```

```python
def first_sections(titles, paragraphs_after_first_step):
    """Собираем индексы глав, включая которые разбиты на несколько параграфов"""
    """На вход список титулов и параграфы после первой обработки, 
    на выходе - первоначальный список индексов глав"""
    sections = []
    for i in range(len(titles)):
        for j in range(len(paragraphs_after_first_step)):
            if titles[i] == paragraphs_after_first_step[j] and (paragraphs_after_first_step[j][:5] == 'Глава' or paragraphs_after_first_step[j][:5] != 'Стать'):
                    sections.append(j)  
            
    return sections
```

```python
def rearrange_sections(first_section,paragraphs_after_first_step):
    """Прикрепляем главы, которые разбиты в одну большую и меняем параграфы"""
    """На вход первоначальный список индексов глав и параграфы после первой обработки,
    на выходе - обработанные параграфы с правильной разметкой глав"""
    sect_list = []
    paragraphs_new = []
    #Находим индексы глав, которые идут подряд (вычленяем главы, разбитые на несколько параграфов)
    for i in range(len(first_section)):
        if first_section[i] - first_section[i-1] == 1:
            sect_list.append(first_section[i-1])
            sect_list.append(first_section[i])
    
    #Сшиваем разбитую главу в одну, удаляя ее прошлые разбитые части
    sect_list = list(set(sect_list))
    for i in range(1,len(sect_list)):
        paragraphs_after_first_step[sect_list[0]] = paragraphs_after_first_step[sect_list[0]]+ ' '+ paragraphs_after_first_step[sect_list[i]]
        paragraphs_after_first_step[sect_list[i]] = ''
    
    #Чистим параграфы от пропусков
    for i in paragraphs_after_first_step:
        if i != '':
            paragraphs_new.append(i)
            
    return paragraphs_new
```

```python
def pick_out_sections(paragraphs_after_first_step):
    """Конечный список индексов глав выводим"""
    """На вход - обработанные параграфы с правильной разметкой глав, 
    на выходе - окончательный список индексов глав"""
    sections = []
    for j in range(len(paragraphs_after_first_step)):
        if paragraphs_after_first_step[j][:5] == 'Глава':
            sections.append(j)
    return sections
```

```python
def pick_out_items(titles,paragraphs_after_first_step):
    """Собираем индексы статей"""
    """На вход - окончательный список титулов и параграфы с правильной разметкой глав, 
    на выходе - список индексов статей"""
    items = []
    for i in range(len(titles)):
        for j in range(len(paragraphs_after_first_step)):
            if titles[i] == paragraphs_after_first_step[j] and paragraphs_after_first_step[j][:5] == 'Стать':
                    items.append(j)   
    return items
```

```python
def make_complicated_items(paragraphs, sections):
    """Собираем индексы статей, если статьи не сделаны обычным образом и сшиты с текстом"""
    """На вход - параграфы,на выходе - список индексов статей"""
    items = []
    for i in range(len(paragraphs)):
        if paragraphs[i][:5] == 'Стать':
            items.append(i)
    
    items = list(reversed(sorted(set(items))))
    
    for j in range(len(paragraphs)):
        if paragraphs[j][:5] != 'Стать' and j != 0 and j not in sections:
            find_st = [j>i for i in items]
            m = find_st.index(True)
            paragraphs[j] = paragraphs[items[m]][:10] +paragraphs[j]
           
    return paragraphs
```

```python
def make_list_without_titles(paragraphs_after_first_step, items,sections):
    """Делаем список индексов -  не названий"""
    """На вход -параграфы с правильной разметкой глав, список индексов статей,
    список индексов глав, на выходе - список индексов параграфов, которые не являются титулами"""
    list_without_titles = []
    for i in range(len(paragraphs_after_first_step)):
        if i not in items and i not in sections:
            list_without_titles.append(i)
    return list_without_titles
```

```python
def make_list_without_titles_no_items(paragraphs_after_first_step,sections):
    list_without_titles = []
    for i in range(len(paragraphs_after_first_step)):
        if i not in sections:
            list_without_titles.append(i)
    return list_without_titles
```

```python
def find_section_and_item_by_index(index,paragraphs,sections,items):
    """Функция для поиска по индексу главы, статьи и части"""
    """На вход - индекс, параграфы обработанные с правильной разметкой, 
    список индексов статей,список индексов глав, 
    на выход - название главы, статьи и номер части"""
    
    items = list(reversed(sorted(set(items))))
    sections = list(reversed(sorted(set(sections))))
    find_gl = [index>i for i in sections]
    find_st = [index>i for i in items]
    
    #Соотносим индекс параграфа и индексы глав и статей
    s = find_gl.index(True)
                
    m = find_st.index(True)
    
    #Проверяем, есть ли номер части или это голая статья
    count = re.compile('\d+.+')
    punct_first = paragraphs[index][:2]
    
    if re.match(count,punct_first) is not None:
        punct = 'ч.' + punct_first
    else:
        punct = ''
        
    section = paragraphs[sections[s]]
    item = paragraphs[items[m]]
    
    return section,item,punct
```

```python
def find_section_and_item_by_index_no_items(index,paragraphs,sections):
    sections = list(reversed(sorted(set(sections))))
    if index !=0:
        find_gl = [index>i for i in sections]
        s = find_gl.index(True)
    
    #Соотносим индекс параграфа и индексы глав
    item = paragraphs[index][:10]            
    
    #Проверяем, есть ли номер части или это голая статья
    count = re.compile('\d+.+')
    punct_first = paragraphs[index][10:12]
    
    if re.match(count,punct_first) is not None:
        punct = 'ч.' + punct_first
    else:
        punct = ''
        
    section = paragraphs[sections[s]]
    
    return section,item,punct
```

```python
def second_step(paragraphs_after_first_step, sections, items, list_without_titles):
    """Пришиваем названия глав к строчкам"""
    """На вход - параграфы обработанные с правильной разметкой, 
    список индексов статей,список индексов глав, список индексов параграфов без титула,
    на выход - параграфы с пришитыми названиями"""
    paragraphs_new_version = []
    
    #Через функцию заданную ранее находим названия глав и пунктов, удаляем номера глав.
    #Пришиваем их к параграфам.
    for i in range(len(paragraphs_after_first_step)):
        if i in sections or i in items:
            paragraphs_new_version.append(paragraphs_after_first_step[i])
        else:
            find = find_section_and_item_by_index(i,paragraphs_after_first_step, sections,items)
            section = find[0][9:]
            item = find[1][10:]
            new = section +' '+ item+' ' + paragraphs_after_first_step[i]
            paragraphs_new_version.append(new)
            
    return paragraphs_new_version
```

```python
def second_step_no_items(paragraphs_after_first_step, sections,list_without_titles):
    paragraphs_new_version = []
    
    for i in range(len(paragraphs_after_first_step)):
        if i in sections:
            paragraphs_new_version.append(paragraphs_after_first_step[i])
        else:
            find = find_section_and_item_by_index_no_items(i,paragraphs,sections)
            section = find[0][9:]
            new = section +' ' + paragraphs_after_first_step[i][10:]
            paragraphs_new_version.append(new)
            
    return paragraphs_new_version
```

```python
def delete_titles(paragraphs, list_without_titles):
    """Убираем все главы и статьи из текста"""
    """На вход - параграфы обработанные с правильной разметкой, 
    список индексов параграфов без титула,
    на выход - параграфы без титулов"""
    
    paragraphs_ready_to_go = []
    for i in range(len(paragraphs)):
        if i in list_without_titles:
            paragraphs_ready_to_go.append(paragraphs[i])
    return paragraphs_ready_to_go
```

```python
def delete_titles_no_items(paragraphs, list_without_titles):
    
    paragraphs_ready_to_go = []
    for i in range(len(paragraphs)):
        if i in list_without_titles:
            paragraphs_ready_to_go.append(paragraphs[i][10:])
    return paragraphs_ready_to_go
```

```python
def process_to_data(doc):
    """Основная функция"""
    """На вход - путь к документу в формате ..docx, на выходе автоматически сохраняются
    docx с пришитыми названиями глав и статей, docx обработанный без пришитых названий,
    словарь размеченный с индексами для определения статей, глав и частей"""
    
    document = Document(doc)
    paragraphs = make_paragraphs_and_list_of_titles(document)
    paragraphs_after_first_step = first_steps(paragraphs)
    titles = list_of_titles(document)
    first = first_sections(titles,paragraphs_after_first_step)
    paragraps_after_first_step = rearrange_sections(first,paragraphs_after_first_step)
    sections = pick_out_sections(paragraphs_after_first_step)
    
    if pick_out_items(titles,paragraphs_after_first_step) == []:
        paragraps_new = make_complicated_items(paragraphs_after_first_step, sections)
        list_without_titles = make_list_without_titles_no_items(paragraphs_after_first_step,sections)
        paragraphs_without_titles = delete_titles_no_items(paragraphs_after_first_step, list_without_titles)
        paragraphs_after_second_step = second_step_no_items(paragraphs_after_first_step, sections,list_without_titles)
        paragraphs_ready_to_go = delete_titles_no_items(paragraphs_after_second_step, list_without_titles)
        dictionary = {}
        for i,idx in enumerate(list_without_titles):
            dictionary[i] = find_section_and_item_by_index_no_items(idx,paragraphs_after_first_step,sections)
    
    else:
        items = pick_out_items(titles,paragraphs_after_first_step)
        list_without_titles = make_list_without_titles(paragraphs_after_first_step, items,sections)
        paragraphs_after_second_step = second_step(paragraphs_after_first_step, sections, items, list_without_titles)
        paragraphs_ready_to_go = delete_titles(paragraphs_after_second_step, list_without_titles)
        paragraphs_without_titles = delete_titles(paragraphs_after_first_step, list_without_titles)
        dictionary = {}
        for i,idx in enumerate(list_without_titles):
            dictionary[i] = find_section_and_item_by_index(idx,paragraphs_after_first_step,sections,items)
    
    name1 = doc+'withtitiles.docx'
    name2 = doc+'without.docx'
    name3 = doc+'.json'
    doc_last = Document()
    for i in paragraphs_ready_to_go:
        doc_last.add_paragraph(i)
    doc_last.save(name1)
    
    doc_new = Document()
    for i in paragraphs_without_titles:
        doc_new.add_paragraph(i)
    doc_new.save(name2)
    

    
    import json
    with open(name3, 'w',encoding='utf8') as fp:
        json.dump(dictionary, fp,ensure_ascii=False)
    
    return 0
```
